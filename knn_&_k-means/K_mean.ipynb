```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r"D:\AI_VIETNAM\file\iris_1D.csv")

X = df.Petal_Length.to_numpy()
X
```

```python
k = 2
# Select random point
C = X[:2]

for _ in range(10):
    distance = np.sqrt((C[:, np.newaxis]-X)**2)  # np.newaxis: +1 kích thước của row/column. Ví dụ mảng ban đầu là (2,) và (6,)
                                                        #after: Mảng 1: $(2, 1) $Mảng 2: $(1, 6)$
    labels = np.argmin(distance, axis=0) # filter ra index min between C0,C1 => select nearby point
    #update point 
    C_new = np.array([X[labels==i].mean() for i in range(k)]) # np.axis ko hỗ trợ list
    if np.all(C_new==C):
        break
    
    C = C_new

# Đánh giá thuật toán
WCSS = np.sum([np.sum( (X[labels== i] - C[i])**2 ) for i in range(k)])
WCSS
```
