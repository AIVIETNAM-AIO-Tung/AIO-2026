```python
C0, C1 = rd.sample(df['Petal_Length'].tolist(), 2)
C0 = pd.Series(C0)
C1 = pd.Series(C1)
print(f"C0: {C0}")
print(f"C1: {C1}")
```

```python
#calculate distance
def distance(C0, C1, X):
    d0 = ((X - C0.values) ** 2)**0.5
    d1 = ((X - C1.values) ** 2)**0.5
    return d0, d1
```

```python
def select_dis(d0, d1, X):
    by_C0 = d0 <= d1
    by_C1 = d0 > d1
    return by_C0, by_C1
```

```python
# Update
def update(C0, C1, iter_num, X):
    for i in range(iter_num):
        C0_old = C0.copy(deep=True) # Nếu ko copy thì C0_old sẽ tham chiếu đến C0
        C1_old = C1.copy(deep=True)

        d0, d1 = distance(C0, C1, X)
        T_C0, T_C1 = select_dis(d0, d1, X)
        C0 = sum(X[T_C0]) / len(X[T_C0])
        C1 = sum(X[T_C1]) / len(X[T_C1])
                
        if (C0_old.values == C0) and (C1_old.values == C1): # Ko thể so sánh đc kiểu Series/Pandas với float
            break                                           # Nhưng kiểu array thì có thể compare vs float
        C0 = pd.Series(C0)
        C1 = pd.Series(C1)
    return C0, C1
```

```python
inter_num = 10
C0, C1 = update(C0, C1, inter_num, X)
C0, C1
```
